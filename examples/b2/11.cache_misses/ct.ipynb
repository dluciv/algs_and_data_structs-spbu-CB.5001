```python
import numpy as np
import numba

REPS = 10
SIZE = 15000
```

```python
@numba.njit
def byrow(reps: int) -> float:
    s: float = 0.0
    m2d = np.zeros((SIZE, SIZE), dtype=float)
    for k in range(reps):
        for i in range(m2d.shape[0]):
            for j in range(m2d.shape[1]):
                s += m2d[i, j]
    return s

@numba.njit
def bycol(reps: int) -> float:
    s: float = 0.0
    m2d = np.zeros((SIZE, SIZE), dtype=float)
    for k in range(reps):
        for i in range(m2d.shape[1]):
            for j in range(m2d.shape[0]):
                s += m2d[j, i]
    return s

# heat
%timeit byrow(1)
%timeit bycol(1)
```

```python
%timeit byrow(REPS)
```

```python
%timeit bycol(REPS)
```
